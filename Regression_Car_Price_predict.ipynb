## Predict car prices based on several features such as Engine HP, Engine Cylinders, and more. Target variable in this dataset is the MSRP

## Step 1: Load and Explore the Dataset

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('https://github.com/RicardoR002/Data/raw/refs/heads/main/CooperUnionDataset.csv')

# Display the first few rows of the dataset
print(df.head())

# Display the column names
print(df.columns)
```

```python
df.info()
```

```python
# check for missing values
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
# check for missing values
df.isnull().sum()
```

## Step 2: Preprocessing the Data

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df.drop(columns=['MSRP'])
y = df['MSRP'].values

# Get encoded feature names and transform data
X_encoded = pd.get_dummies(X, drop_first=True)
feature_names = X_encoded.columns  # Store column names

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
```

## Step 3: Split the Data

```python
from sklearn.model_selection import train_test_split

# Split into train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)
```

## Step 4: Define the Regression Model

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(64, activation='relu', input_shape=(X_scaled.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1)
])
model.compile(optimizer='adam', loss='mean_squared_error')
```

## Step 5: Train the Model

```python
# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)
```

## Step 6: Evaluate the Model

```python
# Evaluate the model
loss = model.evaluate(X_test, y_test, verbose=0)
print('Test Mean Squared Error: %.2f' % loss)
```

## Step 7: Make Predictions

```python
# Make predictions with proper preprocessing
def predict_price(sample_data):
    # Create DataFrame with sample data
    sample_df = pd.DataFrame([sample_data])

    # Encode categorical variables consistently
    sample_encoded = pd.get_dummies(sample_df)

    # Align columns with training data
    for col in feature_names:
        if col not in sample_encoded.columns:
            sample_encoded[col] = 0
    sample_encoded = sample_encoded.reindex(columns=feature_names, fill_value=0)

    # Scale the features
    sample_scaled = scaler.transform(sample_encoded)

    # Make prediction
    prediction = model.predict(sample_scaled)
    return prediction[0][0]

# Example usage
sample_car = {
    'Year': 2011,
    'Engine HP': 335,
    'Engine Cylinders': 6,
    'highway MPG': 26,
    'city mpg': 19,
    'Popularity': 3916,
    'Make': 'BMW',
    'Model': '1 Series M',
    'Engine Fuel Type': 'premium unleaded (required)',
    'Transmission Type': 'MANUAL',
    'Driven_Wheels': 'rear wheel drive',
    'Number of Doors': '2',
    'Market Category': 'Factory Tuner,Luxury,High-Performance',
    'Vehicle Size': 'Compact',
    'Vehicle Style': 'Coupe'
}

# Predict price for the sample car
predicted_price = predict_price(sample_car)
print(f'Predicted MSRP: ${predicted_price:,.2f}')
```
